```python
import streamlit as st
import ollama
```

```python
def load_model():
    host = "http://localhost:11434"  # Change this if your Ollama server is hosted elsewhere
    return ollama.Client(host=host)
```

```python
llm_client = load_model()
```

```python
def main():
    st.title("Sit with Discomfort")
    st.write("This app encourages you to acknowledge and reflect on your discomfort. Take a moment to sit with it.")
```

```python
discomfort = st.text_area("Describe your discomfort:", height=150)
```

```python
model = "llama3.2:3b"
prompt = f"Reflect on this discomfort: {discomfort}. Provide a thoughtful and humorous quote or insight."
generated_response = llm_client.generate(prompt=prompt, model=model, stream=True)
response = ""
with st.empty():
  for chunk in generated_response:
    response += chunk["response"]
    st.write(response)
```
